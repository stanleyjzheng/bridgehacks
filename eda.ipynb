```python
import pandas as pd
```

```python
ob_1 = pd.read_csv('./input/observations_1.csv')
ob_2 = pd.read_csv('./input/observations_2.csv')
ob_3 = pd.read_csv('./input/observations_3.csv')
```

```python
ob_1=(ob_1-ob_1.mean())/ob_1.std()
ob_2=(ob_2-ob_2.mean())/ob_2.std()
ob_3=(ob_3-ob_3.mean())/ob_3.std()
```

```python
ob_1['ob_2_unvax'] = ob_2['infected_unvaccinated']
ob_1['ob_3_unvax'] = ob_3['infected_unvaccinated']
ob_1['ob_2_vax'] = ob_2['infected_vaccinated']
ob_1['ob_3_vax'] = ob_3['infected_vaccinated']
ob_1=(ob_1-ob_1.mean())/ob_1.std()
```

```python
pltr = pd.DataFrame()
pltr['ob_1_comb'] = ob_1['infected_unvaccinated'] + ob_1['infected_vaccinated']
pltr['ob_2_comb'] = ob_1['ob_2_unvax'] + ob_1['ob_2_vax']
pltr['ob_3_comb'] = ob_1['ob_3_unvax'] + ob_1['ob_3_vax']
```

```python
pltr.plot()
```
